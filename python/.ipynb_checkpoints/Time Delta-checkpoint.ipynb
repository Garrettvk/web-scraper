```python
from datetime import datetime



date_str_a = '2020-08-04 23:13:30.449658'
a = datetime.strptime(date_str_a, '%Y-%m-%d %H:%M:%S.%f').date()

date_str_b = '2020-08-04 23:14:45.264627'
b = datetime.strptime(date_str_b, '%Y-%m-%d %H:%M:%S.%f').date()
  

  
# returns a timedelta object 
c = a-b  
print('Difference: ', c) 
  
minutes = c.seconds / 60 if c.seconds >= 60 else c.seconds
print('Difference in minutes: ', minutes) 
```

```python
df = pd.read_csv('../csv/24541 products.csv').fillna('')
df
```

```python
def save_style_number(description): # create a column for style number

    pattern = re.compile(r'\d{6}') # pattern we're looking for
    
    matches = pattern.finditer(description)      # iterate over product descriptions
    
    matches_list = [match.group() for match in matches] # create a list of each match

    return matches_list # return 6 digit style number
```

```python
df['Style Number'] = df['Product Description'].apply(save_style_number) # add style number to dataframe
```

```python
number_of_products = len(df.index)
number_of_products
```

```python
style_number = int(df['Style Number'].loc[0][0])

product_name =  df['Product Name'].loc[0].replace(' ', '-')

product_link = f'https://www.wonatrading.com/product_info.php?products_id={style_number}&kind=2&cPath=172_93_96&description={product_name}'
product_link
```

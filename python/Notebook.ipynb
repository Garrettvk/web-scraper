```python
output_df_list = [] # list cotaining dataframes of each product

for product_link in product_links[0:1]:
    print(product_link) # print current product link
    pages = get_product_links(product_link, driver_1)[0:1] # links for all the products on given page
    df = get_data(pages, driver_1) # dataframe = return of function that scrapes data
    output_df_list.append(df) # add df to list

t1 = time.time() # end time
    
output_df = pd.concat(output_df_list) # combine list into 1 dataframe
output_df # print dataframe
```
